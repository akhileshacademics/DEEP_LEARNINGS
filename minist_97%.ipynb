### 60K images decoded to 784 x 60000 arrays

```python
model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,),activation='sigmoid')
])

model.compile(
    optimizer = 'adam',
    loss= 'sparse_categorical_crossentropy',
    metrics=['accuracy']
    )

model.fit(X_train_flattened, y_train, epochs=5)
```

```python
model_sigmoid = keras.Sequential([
    keras.layers.Dense(128, input_shape=(784,),activation='sigmoid'),
    keras.layers.Dense(10,activation='softmax')
])

model_sigmoid.compile(
    optimizer = 'adam',
    loss= 'sparse_categorical_crossentropy',
    metrics=['accuracy']
    )

model_sigmoid.fit(X_train_flattened, y_train, epochs=5)
```

```python
model.evaluate(X_test_flattened,y_test)
```

```python
plt.matshow(X_test[1])
```

```python
y_pred=model.predict(X_test_flattened)
```

```python
y_pred[1]
```

```python
np.argmax(y_pred[1])
```

```python
y_pred_lable = [np.argmax(i) for i in y_pred]
y_pred_lable[:5]
```

```python
y_test[:5]
```

```python
cm=tf.math.confusion_matrix(labels=y_test,predictions= y_pred_lable)
cm
```

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
