```python
import math
```

```python
def sigmoid(x):
    
    return 1 /(1 + math.exp(-x))
```

```python
sigmoid(0.8)
```

```python
def tanh(x):
    
    return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))
```

```python
tanh(0.44)
```

```python
def relu(x):
    return max(0,x)
```

```python
relu(0.8)
```

```python
def leaky_relu(x):
   return max(0.1*x,x)
```

```python
leaky_relu(2000)
```
